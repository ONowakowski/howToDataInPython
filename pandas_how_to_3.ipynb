#Łączenie danych

```
def fetch_financial_data(company='AMZN'):
  import pandas_datareader.data as pdr
  return pdr.DataReader(name=company, data_source='yahoo')

google = fetch_financial_data('GOOG')
netflix = fetch_financial_data('NFLX')
ups = fetch_financial_data('UPS')
oracle = fetch_financial_data('ORCL')
```

```
google.info()
```

```
netflix.info()
```

```
ups.head()
```

```
oracle.head()
```

```
#zmiana nazw kolumn na rozne, tak aby mozna bylo dodac do jednego df
google.columns = ['google_' + col.lower() for col in google.columns]
netflix.columns = ['netflix_' + col.lower() for col in netflix.columns]
oracle.columns = ['oracle_' + col.lower() for col in oracle.columns]
ups.columns = ['ups_' + col.lower() for col in ups.columns]
```

```
google.head()
```

```
del google['google_adj close']
del netflix['netflix_adj close']
del oracle['oracle_adj close']
del ups['ups_adj close']
```

```
ups.head()
```

Konkatenacja

```
#teraz w jednym df bedzie 20 kolumn, po 5 dla kazdej spolki
df = pd.concat([google, netflix, oracle, ups], axis=1)
df
```

```
#mozemy sformatowac widok
pd.set_option('display.float_format', lambda x: f'{x:.2f}')
df.describe().T
```

```
#wspolczynnik korelacji zmiennych
df.corr()
```

```
#tworzymy liste kolumn z zamknieciami
closes = [col for col in df.columns if col.endswith('close')]
closes
```

```
df[closes].corr()
```

```
#jesli chcemy jeszcze wolumen to:
closes = [col for col in df.columns if col.endswith('close') or col.endswith('volume')]
df[closes]
```
