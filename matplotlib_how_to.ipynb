```
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
sns.set()
```

```
fig = plt.figure()
fig.show()
```

#Wykres liniowy

```
_ = plt.plot([0,1], [0,3])
```

```
plt.plot([0,1], [0,3], label='line', color='green')
plt.xlabel('oś x')
plt.ylabel('oś y')
plt.title('Wykresik przykładowy')
_ = plt.legend()
```

```
_ = plt.plot([0,1,2], [0,3,2])
```

```
x = np.arange(-3, 3, 0.1)
x_sin = np.sin(x)
x_cos = np.cos(x)
plt.plot(x, x_cos)
_ = plt.plot(x, x_sin)
```

#Wiele wykresów

```
fig = plt.figure(figsize=(12,6))
plt.subplot(131) #1 liczba -> wiersze, 2 liczba -> kolumny, 3 liczba -> na ktorym wykresie operuje
plt.plot([0,2,5], [3,3,1])
plt.subplot(132) #wybieram drugi wykres
plt.plot([0,1,2], [20,10, 15], 'r--')
plt.subplot(133)
_ = plt.plot([2,4,6], [2,2,3], 'ro')
```

#Wykresy słupkowe

```
plt.bar(x=[1,2,3], height=[30,40,24])
```

```
plt.bar(x=['a','b','c'], height=[3.5,7.2,9.2]) #dane moga byc tez kategoriami
```

```
plt.barh(y=['a','b','c'], width=[3.5,7.2,9.2]) #wykres horyzontalny
```

#Wykresy punktowe

```
x1 = np.random.randn(100)
x2 = np.random.randn(100)

plt.scatter(x=x1, y=x2)
```

```
#parametr s przypisuje rozmiar do kropek, mozemy wygenerowac losowe rozmiary kazdej z kropek
x3 = np.random.randn(100) * 50

plt.scatter(x=x1, y=x2, s=x3, alpha=0.5) #alpha to przezroczystosc punktow
```
